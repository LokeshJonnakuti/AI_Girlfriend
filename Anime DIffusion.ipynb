```python
# Imports
import torch
from torch import autocast
from diffusers import StableDiffusionPipeline
```

```python
# Used to visualize
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Get the model
pipe = StableDiffusionPipeline.from_pretrained(
    'hakurei/waifu-diffusion',
    torch_dtype=torch.float16,
    cache_dir="D:/python-libs/hugging-face-cache"
).to('cuda')
```

```python
# Remove filter
pipe.safety_checker = lambda images, clip_input: (images, False)
```

```python
# Model details:
# https://gist.github.com/harubaru/f727cedacae336d1f7877c4bbe2196e1
# Settings:
# https://danbooru.donmai.us/wiki_pages/tag_group:image_composition
```

```python
# Create the image
settings = "1girl,solo focus,very wide shot,feamle focus,ratio:16:9,detailed,looking at viewer,facing viewer,facing forward,vtuber"#,simple background,upper body"
characteristics = "waifu,female,brown hair,red eyes,sidelocks,slight,blush,cat ears,neko,maid outfit,shy,slight blush"
# settings = ""
# characteristics = "waifu,female,1girl,brown hair,purple eyes,fox ears,slight blush,upper body,timid,happy,very wide shot,maid"
sent = ""
summary = "'I hope get know better' to viewer"
prompt = f"{settings} {characteristics} {','+sent if len(sent)!=0 else ''}"#, {summary}"
# prompt = f"1girl, very wide shot, looking at viewer, facing viewer, vtuber, black background, chest and head"
with autocast("cuda"):
    image = pipe(prompt, guidance_scale=10)["images"][0]
```
